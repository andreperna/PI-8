```python
# gerar uma lista com 61950 datas de nascimentos aleatórias desde 01/01/1922 até 01/01/2022

import datetime

# data_inicio = datetime.date(1922,1,1)
# data_fim = datetime.date(2022,1,1)

# lista_datas_geral = []
# data_atual = data_inicio
# while data_atual <= data_fim:
#     lista_datas_geral.append(data_atual)
#     data_atual = data_atual + datetime.timedelta(days=1)


# separando algumas faixas etárias
# https://www.duvidasdedireito.com.br/2019/02/crianca-adolescente-jovem-adulto-e-idoso.html
# criança = até 12 anos incompletos         [8.000] 
# adolescente = de 12 até 18 incompletos    [12.000]
# jovem = de 18 até 30 incompletos          [15.000]
# adulto = de 30 até 60 incompletos         [20.000]
# idoso = a partir de 60 anos               [6.950]
# Total ==================================> [61.950]

# função para criar a lista de datas
def retorna_datas(qtd:int, data_inicio:datetime, data_fim:datetime):
    lista_datas = []
    data_atual = data_inicio
    while data_atual <= data_fim:
        lista_datas.append(data_atual)
        data_atual = data_atual + datetime.timedelta(days=1)

    fator_multiplicador = round((qtd / len(lista_datas))+0.5)

    lista_datas = lista_datas * fator_multiplicador
    
    return lista_datas[:qtd]



# criança = até 12 anos incompletos         [8.000]
qtd = 8000
data_inicio = datetime.date(2010,1,1) - datetime.timedelta(days=1)
data_fim = datetime.date(2022,1,1)
datas_criancas = retorna_datas(qtd, data_inicio, data_fim)

# adolescente = de 12 até 18 incompletos    [12.000]
qtd = 12000
data_inicio = datetime.date(2004,1,1) - datetime.timedelta(days=1)
data_fim = datetime.date(2010,1,1)
datas_adolecentes = retorna_datas(qtd, data_inicio, data_fim)

# jovem = de 18 até 30 incompletos          [15.000]
qtd = 15000
data_inicio = datetime.date(1992,1,1) - datetime.timedelta(days=1)
data_fim = datetime.date(2004,1,1)
datas_jovens = retorna_datas(qtd, data_inicio, data_fim)

# adulto = de 30 até 60 incompletos         [20.000]
qtd = 20000
data_inicio = datetime.date(1962,1,1) - datetime.timedelta(days=1)
data_fim = datetime.date(1992,1,1)
datas_adultos = retorna_datas(qtd, data_inicio, data_fim)

# idoso = a partir de 60 anos               [6.950]
qtd = 6950
data_inicio = datetime.date(1922,1,1)
data_fim = datetime.date(1962,1,1)
datas_idosos = retorna_datas(qtd, data_inicio, data_fim)


# lista geral de datas
lista_datas_geral = datas_criancas + datas_adolecentes + datas_jovens + datas_adultos + datas_idosos

# série data_nascimento
serie_data_nasc = pd.Series(lista_datas_geral)
serie_data_nasc

```

```python
# gera uma lista com 61950 numeros aletórios entre 200 e 1200
import random

limites = range(200,1200)
fator_multiplicador = round((91950/len(limites))+0.5)
population = list(limites) * fator_multiplicador


lista_numeros = random.sample(population=population, k=61950)
lista_numeros

# serie
serie_numero = pd.Series(lista_numeros)
serie_numero
```
