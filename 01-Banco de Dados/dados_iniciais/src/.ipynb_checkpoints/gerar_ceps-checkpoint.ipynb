```python
import pandas as pd

df = pd.read_csv('in/ceps.csv', delimiter=';', encoding='utf-8')
display(df.head())

df[['cidade','uf']] = df['cidade-uf'].str.split('/', expand=True)
display(df.head())

df.drop('cidade-uf', axis=1, inplace=True)
display(df.head())

df = df.reindex(columns=['cep','uf','cidade','bairro','logradouro'])
display(df.head())
```
