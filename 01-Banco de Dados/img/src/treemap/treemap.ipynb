```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import squarify
```

```python
# ler csv, criando o DataFrame
df = pd.read_csv('treemap.csv', sep=';')

# classificar em ordem decrescente
df = df.sort_values(by='Cadastros', ascending=False)

# definir as series a serem utilizadas
tamanhos = df['Cadastros']
titulos = df['Escopo']

# definindo mapa de cores
# mapas de cores sequencial como: 'viridis', 'plasma', 'inferno', 'magma', 'Greys', 'Purples', 'Blues' etc. 
mapa_cores = matplotlib.cm.Set1

# normaliza o mapa de cores
valor_maximo = tamanhos.max()
valor_minimo = tamanhos.min()
normalize = matplotlib.colors.Normalize(vmin=valor_minimo, vmax=valor_maximo)

# definindo lista de cores com base no mapa
lista_cores_normalizada = [mapa_cores(normalize(item)) for item in tamanhos]

# divide o mapa de cores em partes iguais
numero_cores = 9
lista_opcoes = list(range(numero_cores))
normalize_dividido = matplotlib.colors.Normalize(vmin=lista_opcoes[0], vmax=lista_opcoes[-1])
lista_cores_dividida = [mapa_cores(normalize_dividido(item)) for item in lista_opcoes]
lista_cores_final = lista_cores_dividida[:5] + lista_cores_dividida[6:7]*4

# definindo lista de cores manual
lista_cores_manual = ['yellowgreen', 'blue', 'green', 'red', 'gray', 'purple', 'purple', 'purple', 'purple']


# plotando o treemap
fig = plt.figure(figsize=(16,8)) 
squarify.plot(sizes=tamanhos, label=tamanhos, color=lista_cores_final, linewidth=1.5, edgecolor="#f5f5f5", alpha=1, text_kwargs={'color':'black', 'weight':'normal', 'size':12})

# # plotando o treemap
# fig, ax = plt.subplots(figsize=(14,8)) 
# squarify.plot(sizes=tamanhos, color=lista_cores_dividida[:5] + lista_cores_dividida[6:7]*4, linewidth=2, edgecolor="#f5f5f5", alpha=1)
# # ax.set_xticks([])
# # ax.set_yticks([])
# ax.plot([0],[0])
# ax.legend(["ssss","ffff","fffwf",])
# red = mpatches.Patch(color='red', label='red')

handles = [mpatches.Patch(color=cor, label=titulo) for cor, titulo in zip(lista_cores_final,titulos)]

plt.legend(handles=handles, fontsize='x-large')
plt.margins(x=None, y=None)
plt.axis('off')
plt.savefig('treemap.png', dpi='figure', bbox_inches='tight', pad_inches=0.03)
plt.show()
```
